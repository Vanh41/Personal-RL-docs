# EP Tối Ưu

Trong phần này, chúng ta sẽ tìm hiểu về các phương pháp tối ưu hóa Evolutionary Programming (EP). EP là một kỹ thuật tối ưu hóa dựa trên quá trình tiến hóa tự nhiên, sử dụng các khái niệm như đột biến, chọn lọc tự nhiên và sinh sản để tìm ra các giải pháp tốt nhất cho các vấn đề phức tạp.

## Các bước cơ bản của EP

1. **Khởi tạo quần thể**: Tạo ra một quần thể ban đầu gồm các cá thể ngẫu nhiên.
2. **Đánh giá**: Đánh giá độ thích nghi của từng cá thể trong quần thể.
3. **Chọn lọc**: Chọn ra các cá thể tốt nhất dựa trên độ thích nghi.
4. **Đột biến**: Áp dụng các phép đột biến để tạo ra các cá thể mới.
5. **Thay thế**: Thay thế các cá thể cũ bằng các cá thể mới.
6. **Lặp lại**: Lặp lại quá trình trên cho đến khi đạt được tiêu chí dừng.

## Ứng dụng của EP

### EP tối thiểu hàm số f(x) = sin(2pi.x).e^-x / [0, 2]

````python
import numpy as np 
import matplotlib.pyplot as plt

def f(x):
    return np.sin(2*np.pi*x)*np.exp(-x)

num_generations = 100
population_size = 20
mutation_sdt = 0.1

# khoi tao 
population = np.random.uniform(0, 2, population_size)
best_fitness_history = []

````

````python
for _ in range(num_generations):
    # danh gia
    fitness = f(population)
    # chon loc
    best_fitness_history.append(np.min(fitness))
    # dot bien
    mutations = np.random.normal(0, mutation_sdt, population_size) 
    offspring = population + mutations
    # gioi han trong khoang [0, 2]
    offspring = np.clip(offspring, 0, 2)
    # chon ca the nho nhat 
    combined_population = np.concatenate((offspring, population))
    combined_fitness = f(combined_population)
    # Chọn lại quần thể bằng cách lấy N cá thể có f(x) nhỏ nhất
    best_indices = np.argsort(combined_fitness)[:population_size]
    population = combined_population[best_indices]
    
# kq cuoi cung
best_x = np.min(population)
best_y = f(best_x)

print(best_x, best_y)
````

````python
# Vẽ biểu đồ quá trình tối ưu
plt.plot(best_fitness_history, label="Best f(x)")
plt.xlabel("Generations")
plt.ylabel("Best f(x)")
plt.title("Quá trình tối ưu f(x) bằng Evolutionary Programming")
plt.legend()
plt.show()
````
